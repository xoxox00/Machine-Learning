```python
#23. Divide the data into input and output y=sales inthousand, x=all other import pandas as pd

# Define 'y' (output variable)
y = df['Sales in thousands']

# Define 'X' (input variables) - All other columns except 'Sales in thousands'
X = df.drop('Sales in thousands', axis=1)

print("Input Features (X):")
print(X.head())

print("\nOutput Variable (y):")
print(y.head())
```

```python
#24.Encode other categorical variables using label encoder
from sklearn.preprocessing import LabelEncoder
import pandas as pd

# Initialize LabelEncoder
label_encoder = LabelEncoder()

# Identify categorical columns (excluding the target column 'Sales in thousands')
categorical_cols = X.select_dtypes(include=['object']).columns

# Encode each categorical column
for col in categorical_cols:
    X[col] = label_encoder.fit_transform(X[col])

print("Encoded DataFrame (X):")
print(X.head())
```

```python
#25. Encode categorical variable vehicle type using one-hot encoder
import pandas as pd
from sklearn.preprocessing import OneHotEncoder

# Initialize OneHotEncoder with updated syntax
encoder = OneHotEncoder(sparse_output=False, drop='first')  # Correct parameter

# Encode 'Vehicle type'
vehicle_encoded = encoder.fit_transform(df[['Vehicle type']])

# Convert encoded array back to DataFrame
encoded_df = pd.DataFrame(vehicle_encoded, columns=encoder.get_feature_names_out(['Vehicle type']))

# Combine with original DataFrame
df = pd.concat([df, encoded_df], axis=1)

# Drop original 'Vehicle type' column (optional)
df.drop('Vehicle type', axis=1, inplace=True)

print(df.head())

```

```python
#26..Split the data set into train and test set 70% train set 10% test set
from sklearn.model_selection import train_test_split

# Split the data into train (70%) and test (30%)
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)

# Display the shapes of the resulting datasets
print(f"X_train shape: {X_train.shape}")
print(f"X_test shape: {X_test.shape}")
print(f"y_train shape: {y_train.shape}")
print(f"y_test shape: {y_test.shape}")
```

```python
#27. Apply feature scaling on numerical variables
from sklearn.preprocessing import StandardScaler

# Initialize StandardScaler
scaler = StandardScaler()

# Identify numerical columns for scaling
numerical_cols = X.select_dtypes(include=['int64', 'float64']).columns

# Apply scaling
X[numerical_cols] = scaler.fit_transform(X[numerical_cols])

print("Scaled DataFrame:")
print(X.head())
```
