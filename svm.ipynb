```python
x=df.iloc[:,:-1]
y=df["Species"]
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test = train_test_split(x,y,test_size=0.3,random_state=1)
print(x_train,y_train)
```

```python
from sklearn.svm import SVC
sv=SVC(kernel="linear")
sv.fit(x_train,y_train)
sv.score(x_test,y_test)*100,sv.score(x_train,y_train)*100
```

```python
sv.predict([[5.2,2.7,3.9,1.4]])
```

```python
from sklearn.metrics import confusion_matrix
cf = confusion_matrix(y_test,sv.predict(x_test))
cf
```

```python
sns.heatmap(cf,annot=True)
plt.show()
```

```python
from sklearn.metrics import accuracy_score,classification_report,f1_score,precision_score,recall_score
```

```python
cr=classification_report(y_test,sv.predict(x_test))
print(cr)
```

```python
accuracy = accuracy_score(y_test,sv.predict(x_test))
accuracy*100
```

```python
f1score = f1_score(y_test,sv.predict(x_test),average="macro")
f1score*100
```

```python
precision = precision_score(y_test,sv.predict(x_test),average="macro")*100
precision
```

```python
recall = recall_score(y_test,sv.predict(x_test),average="macro")*100
recall
```
