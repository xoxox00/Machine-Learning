```python
x = df["TV"]
y = df["Sales"]
```

```python
from sklearn.model_selection import train_test_split
```

```python
x_train,x_test,y_train,y_test = train_test_split(x,y,test_size=0.2,random_state=42)
```

```python
x_mean,y_mean = np.mean(x_train),np.mean(y_train)
numerator=np.sum((x_train.values.flatten()-x_mean)*(y_train.values-y_mean))
denominator=np.sum((x_train.values.flatten()-x_mean)**2)
print(numerator,denominator)
b1=numerator/denominator
print(b1)
b0=(y_mean-(b1*x_mean))
print(b0)
print("x mean:",x_mean)
print("y mean:",y_mean)
```

```python
print(f"model is y=:{b0}+{b1}x")
```

```python
X = np.array(df["TV"]).reshape(-1,1)
Y = np.array(df['Sales'])
```

```python
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
x1_train,x1_test,y1_train,y1_test = train_test_split(X,Y,test_size=0.2,random_state=42)
lr = LinearRegression()
lr.fit(x1_train,y1_train)
print("score:",lr.score(x1_test,y1_test)*100)
print(lr.coef_,lr.intercept_)
```
